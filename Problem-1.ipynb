# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
def random_1D(step):
    trajectory = np.zeros(step)
    for i in range(step-1):
        trajectory[i+1] = trajectory[i] + np.random.random() - 1/2
    return trajectory
def EoT(step, ensemble_step):
    ensemble = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_1D(step)
        ensemble[i] = ram[step-1]
    return ensemble
```

```python
Tstep = 100
test1 = random_1D(Tstep)
T = np.linspace(0, Tstep, np.size(test1))
fig, ax = plt.subplots(figsize=(5, 5))
plt.bar(T, test1)
plt.show()
```

```python
ensemble_test = EoT(100,100000)
mu, sigma = 0, 3
fig, ax = plt.subplots(figsize=(8,8))
_ = ax.hist(ensemble_test, 50, density=True)
x = np.linspace(-10, 10)
ax.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
```

```python
def random_1D_2(step):
    trajectory = np.zeros(step)
    for i in range(step-1):
        flag = np.random.random()
        if(flag < 1/2):
            trajectory[i+1] = trajectory[i] - 1
        elif(flag >= 1/2):
            trajectory[i+1] = trajectory[i] + 1
    return trajectory
def EoT_2(step, ensemble_step):
    ensemble = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_1D_2(step)
        ensemble[i] = ram[step-1]
    return ensemble
```

```python
Tstep = 100
test2 = random_1D_2(Tstep)
T = np.linspace(0, Tstep, np.size(test2))
fig, ax = plt.subplots(figsize=(5, 5))
plt.bar(T, test2)
plt.show()
```

```python
ensemble_test_2 = EoT_2(50,200000)
```

```python
mu, sigma = 0, 7
fig, ax = plt.subplots(figsize=(8,8))
_ = ax.hist(ensemble_test_2, 30, density=True)
x = np.linspace(-20, 20)
ax.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
```
