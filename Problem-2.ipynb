# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
def random_2D(step):
    trajectory_x = np.zeros(step)
    trajectory_y = np.zeros(step)
    for i in range(step-1):
        trajectory_x[i+1] = trajectory_x[i] + np.random.random() - 1/2
        trajectory_y[i+1] = trajectory_y[i] + np.random.random() - 1/2
    return trajectory_x, trajectory_y
def EoT(step, ensemble_step):
    ensemble_x = np.zeros(ensemble_step)
    ensemble_y = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D(step)
        ensemble_x[i] = ram[0][step-1]
        ensemble_y[i] = ram[1][step-1]
    return ensemble_x, ensemble_y

def EoT_square(step, ensemble_step):
    ensemble = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D(step)
        ensemble[i] = ram[0][step-1]**2 + ram[1][step-1]**2
    return ensemble
```

```python
Tstep = 100
test1 = random_2D(Tstep)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(test1[0], test1[1])
ax.plot(test1[0][Tstep-1], test1[1][Tstep-1], c='red', marker='+', label = 'end point')
ax.legend()
plt.show()
```

```python
ensemble_test = EoT(50,200000)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
_ = ax.hist2d(ensemble_test[0], ensemble_test[1], [30, 30])
plt.show()
```

```python
def random_2D_2(step):
    trajectory_x = np.zeros(step)
    trajectory_y = np.zeros(step)
    
    for i in range(step-1):
        flag_x = np.random.random()
        flag_y = np.random.random()
        if(flag_x < 1/2):
            trajectory_x[i+1] = trajectory_x[i] - 1
        elif(flag_x >= 1/2):
            trajectory_x[i+1] = trajectory_x[i] + 1
        if(flag_y < 1/2):
            trajectory_y[i+1] = trajectory_y[i] - 1
        elif(flag_y >= 1/2):
            trajectory_y[i+1] = trajectory_y[i] + 1  
    return trajectory_x, trajectory_y
def EoT_2(step, ensemble_step):
    ensemble_x = np.zeros(ensemble_step)
    ensemble_y = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D_2(step)
        ensemble_x[i] = ram[0][step-1]
        ensemble_y[i] = ram[1][step-1]
    return ensemble_x, ensemble_y

def EoT_square_2(step, ensemble_step):
    ensemble = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D_2(step)
        ensemble[i] = ram[0][step-1]**2 + ram[1][step-1]**2
    return ensemble
```

```python
Tstep = 100
test2 = random_2D_2(Tstep)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(test2[0], test2[1])
ax.plot(test2[0][Tstep-1], test2[1][Tstep-1], c='red', marker='+', label = 'end point')
ax.legend()
plt.show()
```

```python
ensemble_test2 = EoT_2(50,200000)
fig, ax = plt.subplots(figsize=(5,5))
_ = ax.hist2d(ensemble_test2[0], ensemble_test2[1], [30, 30])
plt.show()
```

```python
def random_2D_3(step):
    trajectory_x = np.zeros(step)
    trajectory_y = np.zeros(step)
    
    for i in range(step-1):
        theta = 2.0*np.pi*np.random.random()
        trajectory_x[i+1] = trajectory_x[i] + np.cos(theta)
        trajectory_y[i+1] = trajectory_y[i] + np.sin(theta)  
    return trajectory_x, trajectory_y
def EoT_3(step, ensemble_step):
    ensemble_x = np.zeros(ensemble_step)
    ensemble_y = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D_3(step)
        ensemble_x[i] = ram[0][step-1]
        ensemble_y[i] = ram[1][step-1]
    return ensemble_x, ensemble_y

def EoT_square_3(step, ensemble_step):
    ensemble = np.zeros(ensemble_step)
    for i in range (ensemble_step):
        ram = random_2D_3(step)
        ensemble[i] = ram[0][step-1]**2 + ram[1][step-1]**2
    return ensemble
```
